## Analyze gender distribution

### Subtask:
Analyze the distribution of genders in the dataset and visualize it using a chart.


**Reasoning**:
Map the GenderId to meaningful labels, count the occurrences, and create a bar plot to visualize the gender distribution.


```python
df['Gender'] = df['GenderId'].map({1: 'Male', 2: 'Female'})
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6, 4))
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Distribution of Genders in the Dataset')
plt.xlabel('Gender')
plt.ylabel('Number of Clients')
plt.show()
plt.savefig('gender_distribution.png')
```

## Analyze bank deposits by gender

### Subtask:
Analyze the average bank deposits by gender and visualize it using a chart.


**Reasoning**:

```python
average_deposits_by_gender = df.groupby('Gender')['Bank Deposits'].mean().reset_index()

plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Bank Deposits', data=average_deposits_by_gender)
plt.title('Average Bank Deposits by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Bank Deposits')
plt.show()
plt.savefig('average_bank_deposits_by_gender.png')
```

## Analyze banking contact by loyalty classification

### Subtask:
Analyze the distribution of banking contacts for each loyalty classification and visualize it using a chart.

**Reasoning**:
Group the DataFrame by 'Loyalty Classification' and 'Banking Contact', count the occurrences, and create a stacked bar plot to visualize the distribution of banking contacts within each loyalty classification.

```python
banking_contact_loyalty = df.groupby(['Loyalty Classification', 'Banking Contact']).size().unstack(fill_value=0)

banking_contact_loyalty.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Distribution of Banking Contacts by Loyalty Classification')
plt.xlabel('Loyalty Classification')
plt.ylabel('Number of Clients')
plt.xticks(rotation=0)
plt.legend(title='Banking Contact', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
plt.savefig('banking_contact_by_loyalty_classification.png')
```
